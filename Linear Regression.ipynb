```python
# PREPROCESSING: Clean dataset (keep duplicates, remove nulls and empty column)
import pandas as pd

# Load data
data = pd.read_excel("input_to_linear_regression2.xlsx")

# 1. Drop Unnamed: 8 column (100% empty)
if "Unnamed: 8" in data.columns:
    data = data.drop(columns=["Unnamed: 8"])

# 2. Remove rows with null/NA values (duplicates are KEPT as per project requirement)
data = data.dropna()

print(f"Cleaned shape: {data.shape[0]} rows x {data.shape[1]} columns")
print(f"Null values remaining: {data.isnull().sum().sum()}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Use preprocessed data from cell 0 (already cleaned: nulls removed, Unnamed: 8 dropped)

# features (independent variables)
X = data[[
    "1-syllable_words",
    "2-syllable_words",
    "3-syllable_words",
    "4-syllable_words",
    "5-syllable_words",
    "6-syllable_words",
    "7-syllable_words",
    "8-syllable_words",
    "word_length-1",
    "word_length-2",
    "word_length-3",
    "word_length-4",
    "word_length-5",
    "word_length-6",
    "word_length-7",
    "word_length-8",
    "sentence_length"
]]

# target variable
y = data["avg_syllable/avg_word_length"]

# split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# create model
model = LinearRegression()

# train model
model.fit(X_train, y_train)

# prediction
predictions = model.predict(X_test)

# print results
print("Predictions:", predictions)

# model coefficients
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# accuracy
score = model.score(X_test, y_test)
print("R2 Score:", score)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Use preprocessed data from cell above (already cleaned)

# Generate random grade level (for example 1–12)
np.random.seed(42)   # for reproducibility
data["grade_level"] = np.random.randint(1, 13, size=len(data))

# Features (independent variables)
X = data[[
    "1-syllable_words",
    "2-syllable_words",
    "3-syllable_words",
    "4-syllable_words",
    "5-syllable_words",
    "6-syllable_words",
    "7-syllable_words",
    "8-syllable_words",
    "word_length-1",
    "word_length-2",
    "word_length-3",
    "word_length-4",
    "word_length-5",
    "word_length-6",
    "word_length-7",
    "word_length-8",
    "sentence_length",
    "avg_syllable/avg_word_length",
    "average_word_length/sentence_length"

]]

# Target variable
y = data["grade_level"]

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create Linear Regression model
model = LinearRegression()

# Train model
model.fit(X_train, y_train)

# Predict
predictions = model.predict(X_test)

# Print results
print("Predicted grade levels:", predictions)

# Model score
print("R² Score:", model.score(X_test, y_test))


data.to_excel("dataset_with_grade.xlsx", index=False)
```
